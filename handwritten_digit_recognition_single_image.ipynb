```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
image=cv.imread("mnist.jpg")
print(image.shape)
```

```python
gray=cv.cvtColor(image, cv.COLOR_BGR2GRAY)
print(gray.shape)
resize=cv.resize(gray,(28,28))
print(resize.shape)
```

```python
cv.imshow("mnist image",gray)
cv.imshow("resize image",resize)
cv.imshow("original image",image)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
# Convert image to 1D feature vector
features = resize.flatten()
print(features)

# Normalize
features = features / 255.0
print(features)
```

```python
X_train = features.reshape(1, -1)  # 1 sample, 784 features
y_train = [9]  # label
```

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=1)

model.fit(X_train, y_train)
```

```python
prediction = model.predict(X_train)
print("Predicted Digit:", prediction[0])
```
